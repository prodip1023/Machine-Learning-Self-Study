<div style="text-align: center; color: orange; font-size: 24px; font-weight: bold;">
    THE DUMMY VARIABLES METHOD
</div>

```python
get_dummies = pd.get_dummies(homeprice['town'],dtype='int')
get_dummies
```

```python
get_dummies = get_dummies.drop(['west windsor'],axis='columns')
get_dummies
```

- Let us add these get_dummies variables to the original data frame using concat() method of pandas package.

```python
merged = pd.concat([homeprice,get_dummies],axis='columns')
merged
```

- In the above output,we do not require 'town' column as it is already replaced by the get_dummies variables (the last two columns).Hence let us the 'town' column.

```python
final = merged.drop(['town'],axis='columns')
final
```

- This 'final' dataset should be used for analysis purpose.
- Now divided independent variables (x) and dependent variables(y).

```python
x = final.drop(['price'],axis='columns')
y = final['price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2800,1,0]]) # Moreno township
```

```python
model.predict([[2800,0,1]]) # Robinsvile 
```

```python
model.predict([[2800,0,0]]) # Neither Moreno township and Robinsvile
```

```python
model.score(x,y)
```

```python
df = homeprice.copy()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.town = le.fit_transform(df.town)
df
```

```python
x = df[['town','area']].values
y = df['price']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
x1 = ohe.fit_transform(df[['town']]) 
x1 = pd.DataFrame(x1.toarray())
x1
```

```python
x1 = x1.iloc[:,1:]
x1
```

```python
x = pd.concat([df, x1],axis='columns')
x
```

```python
x.drop(['town','price'],axis='columns',inplace=True)
```

```python
x
```

```python
x = x.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2800,1,0]])
```
