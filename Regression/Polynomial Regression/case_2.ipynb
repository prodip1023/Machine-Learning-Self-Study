```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/Regression/Polynomial Regression/Data/Salary_Experience.csv')
```

```python
# First 5 rows
df.head()
```

```python
# Retrieve level the col, i.e 1st column as 2D array
x = df.iloc[:,:1].values
x
```

```python
# Retrieve Salary Col i.e 2nd column as 1D array
y = df.iloc[:,1].values
y
```

```python
# Let us Scatter plot the distribution of the data points
plt.scatter(x,y,color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience')
plt.show()
```

```python
# Create the Polynomial Features with degree 5
pf = PolynomialFeatures(degree=5)
```

```python
# Convert x into Polynomial regression with degree 5
x1 = pf.fit_transform(x)
```

```python
# Apply polynomial linear regression ox x1 and y
lr = LinearRegression()
lr.fit(x1,y)
```

```python
# observe the r squared value .Best fit if it is closer to 1
print("Polynomial Regression Model accuracy:",lr.score(x1,y))
```

```python
# Draw data points along with regression line
plt.scatter(x,y)
plt.plot(x,lr.predict(x1))
plt.show()
```

```python
# Find the coefficient
lr.coef_
```

```python
# Find the intercept
lr.intercept_
```

```python
# Judge the salary of an employee having 5.5 years of experience.
inputs = [[5.5]]
inputs_1 = pf.fit_transform(inputs)
result = lr.predict(inputs_1)
print('Salary for 5.5 years of experience:', result)
```
