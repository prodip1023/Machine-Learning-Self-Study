```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import numpy as np
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/NAIVE_BAYES_CLASSIFIACTION/Data/cricket.csv')
```

```python
categorical_value = ['OUTLOOK','TEMPERATURE','HUMIDITY','WINDY','PLAY CRICKET']
```

```python
for i in range(0,len(categorical_value)):
    print(data[categorical_value[i]].unique())
    print(data[categorical_value[i]].value_counts())
```

- The 4 feature columns in the dataset contain categorical data.
- The Machine learning models will not work directly on categorical values.They work only numerical type of data.
- Hence we have to convert categorical values into numeric type of data.
- We can use dummy variables method or one hot encoding method.Not use here these type of methods.
- We create own functions to convert categorical values to numeric values.

```python
def outlook_to_numeric(x):
    if x == 'Rainy':
        return 0
    if x == 'Overcast':
        return 1
    if x == 'Sunny':
        return 2
```

```python
data['OUTLOOK'] = data['OUTLOOK'].apply(outlook_to_numeric)
```

```python
data.head()
```

```python
def temp_to_numeric(x):
    if x == 'Hot':
        return 0
    if x == 'Mild':
        return 1
    if x == 'Cool':
        return 1
```

```python
data['TEMPERATURE'] = data['TEMPERATURE'].apply(temp_to_numeric)
```

```python
data.head()
```

```python
def humid_to_numeric(x):
    if x == 'High':
        return 0
    if x == 'Normal':
        return 1
```

```python
data['HUMIDITY'] = data['HUMIDITY'].apply(humid_to_numeric)
```

```python
data.head()
```

```python
def windy_to_numeric(x):
    if x == False:
        return 0
    if x == True:
        return 1
```

```python
data['WINDY'] = data['WINDY'].apply(windy_to_numeric)
```

```python
data.head()
```

```python
def target_to_numeric(x):
    if x == 'No':
        return 0
    if x == 'Yes':
        return 1
```

```python
data['PLAY CRICKET'] = data['PLAY CRICKET'].apply(target_to_numeric)
```

```python
data.head()
```

```python
# Independent and dependent features
X = data[['OUTLOOK', 'TEMPERATURE', 'HUMIDITY', 'WINDY']]
y = data['PLAY CRICKET']
```

```python
# train and test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=10)
```

```python
# Gaussian Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
gnb_pred = gnb.predict(X_test)
```

```python
# Check the accuracy of the model
from sklearn import metrics
print("Gaussian Naive Bayes Accuracy:",metrics.accuracy_score(y_test, gnb_pred))
print("Accuracy :",gnb.score(X_test,y_test))
```

```python
# give the new data on that model and predict play cricket or not
''' 
OUTLOOK -> Rainy-0,Overcast-1,Sunny-2
TEMPERATURE -> Hot-0,Mild-1,Cool-1
HUMIDITY -> High-0,Normal-1
WINDY -> False-0,True-1
PLAY CRICKET(target) -> No-0,Yes-1
'''
gnb.predict([[0,1,0,1]])
```
