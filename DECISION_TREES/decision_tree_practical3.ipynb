# DECISION TREE CLASSIFIER PRACTICAL IMPLEMENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# import iris dataset using sklearn
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# independent Features
X= pd.DataFrame(iris['data'],columns=iris.feature_names)
```

```python
# dependent feature
y = iris['target']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=10)
```

```python
# No need to require Standarization or normalization
# apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
# visualize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(dtree, feature_names=iris.feature_names, filled=True, rounded=True)
plt.show()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtree,filled=True)
```

```python
# Post pruning
from sklearn.tree import DecisionTreeClassifier
dtree_post_pruning = DecisionTreeClassifier(max_depth=2)
dtree_post_pruning.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtree_post_pruning,filled=True)
```

```python
y_pred = dtree_post_pruning.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm = classification_report(y_test,y_pred)
print(cm)
print(confusion_matrix(y_test,y_pred))
```

```python
parms = {
    'criterion': ['gini', 'entropy','log_loss'],
    'splitter': ['best','random'],
    'max_depth': [ 1, 2, 3, 4, 5],
    'max_features':['auto','sqrt','log2']
}
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# gridsearchcv
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid=parms, cv=5,scoring='accuracy',verbose=3)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
y_pred = grid_search.predict(X_test)
cm = classification_report(y_test,y_pred)
print(cm)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
