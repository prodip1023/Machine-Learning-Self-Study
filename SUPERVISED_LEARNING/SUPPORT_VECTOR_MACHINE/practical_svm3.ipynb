# SUPPORT VECTOR CLASSIFIER IMPLEMENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

##### Lets create synthetic data points

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue= y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc_linear = SVC(kernel='linear')
```

```python
svc_linear.fit(X_train,y_train)
```

```python
# Prediction
y_pred = svc_linear.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(svc_linear.score(X_test,y_test))
```

```python
svc_linear.intercept_
```

```python
svc_linear.coef_
```

```python
svc_poly = SVC(kernel='poly')
```

```python
svc_poly.fit(X_train,y_train)
```

```python
y_pred_poly = svc_poly.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_poly))
print(confusion_matrix(y_test,y_pred_poly))
print(svc_poly.score(X_test,y_test))
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue= y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(svc.score(X_test,y_test))
```

```python
svc_rbf = SVC(kernel='rbf')
```

```python
svc_rbf.fit(X_train,y_train)
```

```python
y_pred = svc_rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(svc_rbf.score(X_test,y_test))
```

```python
svc_poly = SVC(kernel='poly')
```

```python
svc_poly.fit(X_train,y_train)
```

```python
y_pred_poly = svc_poly.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_poly))
print(confusion_matrix(y_test,y_pred_poly))
print(svc_poly.score(X_test,y_test))
```

```python
svc_sigmoid = SVC(kernel='sigmoid')
svc_sigmoid.fit(X_train,y_train)
y_pred_sigmoid = svc_sigmoid.predict(X_test)
print(classification_report(y_test,y_pred_sigmoid))
print(confusion_matrix(y_test,y_pred_sigmoid))
print(svc_sigmoid.score(X_test,y_test))
```

#### Hyperparameter tuning in SVM

```python
from sklearn.model_selection import GridSearchCV
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred= grid.predict(X_test)
print(classification_report(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
print(grid.score(X_test,y_test))
```
