```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/SUPERVISED_LEARNING/REGRESSION/Polynomial Regression/Data/Ice-cream-selling-data.csv')
df.head()
```

```python
plt.scatter(df['Temperature (°C)'],df['Ice Cream Sales (units)'])
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
x =df[['Temperature (°C)']].values
y = df['Ice Cream Sales (units)'].values
```

```python
poly_model = make_pipeline(PolynomialFeatures(degree=3),LinearRegression())
poly_model.fit(x,y)
```

```python
y_pred = poly_model.predict(x)
```

```python
plt.scatter(df['Temperature (°C)'],df['Ice Cream Sales (units)'])
plt.plot(x,y_pred,color='red')
```

```python
poly_model_3 = make_pipeline(PolynomialFeatures(degree=3),LinearRegression())
poly_model_3.fit(x,y)
poly_model_4 = make_pipeline(PolynomialFeatures(degree=4),LinearRegression())
poly_model_4.fit(x,y)
poly_model_5 = make_pipeline(PolynomialFeatures(degree=5),LinearRegression())
poly_model_5.fit(x,y)
poly_model_6 = make_pipeline(PolynomialFeatures(degree=6),LinearRegression())
poly_model_6.fit(x,y)
poly_model_7 = make_pipeline(PolynomialFeatures(degree=7),LinearRegression())
poly_model_7.fit(x,y)
```

```python
plt.scatter(df['Temperature (°C)'],df['Ice Cream Sales (units)'])
plt.plot(x,poly_model_3.predict(x),color='red')
plt.plot(x,poly_model_4.predict(x),color='blue')
plt.plot(x,poly_model_5.predict(x),color='orange')
plt.plot(x,poly_model_6.predict(x),color='yellow')
plt.plot(x,poly_model_7.predict(x),color='black')
```
