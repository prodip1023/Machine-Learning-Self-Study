```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/Regression/Ridge_Regression/Data/train.csv')
df_test = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/Regression/Ridge_Regression/Data/test.csv')
```

```python
df_test.head()
```

```python
w_train = df_train.iloc[:,0:1].values
h_train = df_train.iloc[:,1].values
```

```python
w_test = df_test.iloc[:,0:1].values
h_test = df_test.iloc[:,1].values
```

- Let us draw a scatter plot for train and test data in the same graph

```python
plt.scatter(w_train,h_train,color='blue',marker='o')
plt.scatter(w_test,h_test,color='red',marker='d')
plt.show()
```

- We can use Simple Linear Regression for train data and Ridge Regression for test data.
- To apply Ridge Regression Model,first we should create an object to Ridge class in linear_model module of sklearn package.
- https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.Ridge.html

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=0.001)
# lambda = 0.001,3,2,1,0.5,05,0.01,0.005 etc and findout out where the accuracy is better.
```

```python
# To train out Ridge Regression Model on the test data,we can use fit() method
rr.fit(w_test,h_test)
```

- The score can be obtained by calling score() method

```python
rr.score(w_test,h_test)
```

```python
rr_1 = Ridge(alpha=0.5)
rr_1.fit(w_test,h_test)
rr_1.score(w_test,h_test)
```

- In these case,we may get a score of 88%.
- the hyper parameter that is tuned is "alpha" value.

```python
rr.predict([[70.3]])
```
