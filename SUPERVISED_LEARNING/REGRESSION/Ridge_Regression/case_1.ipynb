1. We have weight and height and of some people of in 2 seperate .csv files.Use train.csv and test.csv files to retrieve training and testing data.Fit Simple Linear Regression on training data and Ridge Regression on testing data.Also,predict height of a person having 70.3 kg weight.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LinearRegression,Ridge
```

```python
# Load training data into x and test data into y
x = pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/Regression/Ridge_Regression/Data/train.csv')
y =  pd.read_csv('/Users/hackthebox/Downloads/Machine-Learning-Self-Study/Regression/Ridge_Regression/Data/test.csv')
```

```python
# take the data into x and y variables
w_train = x.iloc[:,0:1].values
h_train = x.iloc[:,1].values
w_test = y.iloc[:,0:1].values
h_test = y.iloc[:,1].values
```

```python
# draw scatter plot for train and test data in the same graph
plt.scatter(w_train,h_train,color='blue',marker='o')
plt.scatter(w_test,h_test,color='red',marker='d')
plt.show()
```

```python
# Using Simple Linear Regression on fit train data
lr = LinearRegression()
lr.fit(w_train,h_train)
```

```python
# find the score of the Simple Linear Regression on train data
lr.score(w_train,h_train)
```

```python
# draw scatter plot with orginal data and predicted line 
plt.scatter(w_train,h_train,color='blue',marker='o')
plt.plot(w_train,lr.predict(w_train),color='red')
plt.show()
```

```python
# Using Ridge Regression on fit a line for test data
rr = Ridge(alpha=0.001)
rr.fit(w_test,h_test)
```

```python
# find the score of the Ridge Regression on test data
rr.score(w_test,h_test)
```

```python
# draw scatter plot with orginal data and predicted line 
plt.scatter(w_test,h_test,color='red')
plt.plot(w_test,lr.predict(w_test),color='green')
plt.show()
```

```python
# predict height of the person having 70.3 kg weight
rr.predict([[70.3]])
```
