```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
x = iris.data
y = iris.target
```

```python
df = pd.DataFrame(x,columns=iris.feature_names)
```

```python
df['targets'] = y
```

```python
df
```

```python
scaler = StandardScaler()
x_scaled =scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.2,random_state=43)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(x_train,y_train)
```

```python
knn_model.predict(x_test)
```

```python
print(accuracy_score(y_test,knn_model.predict(x_test)))
print(confusion_matrix(y_test,knn_model.predict(x_test)))
print(classification_report(y_test,knn_model.predict(x_test)))
```

```python
params = {
    "n_neighbors":np.arange(0,100),
}
grid_knn = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=params, cv=5,scoring='accuracy',n_jobs=1,verbose=True)
grid_knn.fit(x_train,y_train)
```

```python
grid_knn.best_estimator_
```

```python
grid_knn.best_params_
```

```python
knn_model_update = KNeighborsClassifier(n_neighbors=9)
knn_model_update.fit(x_train,y_train)
```

```python
knn_model_update.predict(x_test)
```

```python
print(accuracy_score(y_test,knn_model_update.predict(x_test)))
print(confusion_matrix(y_test,knn_model_update.predict(x_test)))
print(classification_report(y_test,knn_model_update.predict(x_test)))
```
