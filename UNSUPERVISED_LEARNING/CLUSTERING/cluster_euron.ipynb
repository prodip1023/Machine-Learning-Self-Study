```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x,y = make_blobs(n_samples=300,centers=4,cluster_std=.60,random_state=30)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
kmeans= KMeans(n_clusters=4,init='k-means++',random_state=30)
kmeans
```

```python
y_means_predict = kmeans.fit_predict(x)
```

```python
cents = kmeans.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y_means_predict)
plt.scatter(cents[:,0],cents[:,1],c='red',marker='*')
```

```python
inertia = [] # wcss
k = range(1,11)
for i in k:
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=30)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(k,inertia,marker='o')
```

```python
kmeans.predict([[.4545,.3456]])
```

```python
silhouette_score(x,y_means_predict) # if k= 3 score is better 
```
