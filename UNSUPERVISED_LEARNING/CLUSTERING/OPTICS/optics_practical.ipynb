```python
from sklearn.cluster import OPTICS
from sklearn import datasets
import numpy as np

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data

# Fit the OPTICS model
clustering = OPTICS(min_samples=20, xi=.05, min_cluster_size=.05)
clustering.fit(X)

# Extract the clusters
labels = clustering.labels_

# Print the cluster labels
print("Cluster Labels:", labels)
```

```python
X
```

```python
# validate the cluster for using Silhoutte score
from sklearn.metrics import silhouette_score
silhouette_score(X,clustering.labels_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,2],c=clustering.labels_)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.title("OPTICS Clustering on IRIS Dataset")
plt.show()
```
