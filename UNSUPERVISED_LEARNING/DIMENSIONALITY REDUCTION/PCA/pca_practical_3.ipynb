1. Take the scikit learn breast cancer data set with 30 dimensions.Reduced the dimensions to 2D using PCA.Then using Naive Bayes model,classify the cancer of a patient into malignant or benign

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
cancer = load_breast_cancer()
dir(cancer)
```

```python
print(cancer.DESCR)
```

```python
cancer.data
```

```python
cancer.feature_names
```

```python
# take independent features as x
x = pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
x
```

```python
# Take dependent feature as y 
y = cancer.target
```

```python
y
```

```python
# create normalized the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Using PCA class to reduce 2D dimensions
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_projected = pca.fit_transform(x_scaled)
x_projected
```

```python
pc1 = x_projected[:,0]
pc2 = x_projected[:,1]
```

```python
# Create dataframe
df = pd.DataFrame(x_projected,columns=['PC1','PC2'])
df['target'] = y
df
```

```python
import seaborn as sns
sns.scatterplot(data = df,x = 'PC1',y = 'PC2',hue='target')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
df = df.drop('target',axis='columns')
df.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df,y,test_size=0.2)
```

```python
# Using Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
