```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
```

```python
df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
df.head()
```

```python
x = df.drop(['Performance Index'],axis=1)
y = df['Performance Index']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
scaler = StandardScaler()
```

```python
x_train_sd = scaler.fit_transform(x_train)
x_test_sd = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train_sd,y_train)
```

```python
y_pred = lr_model.predict(x_test_sd)
```

```python
x_test_sd
```

```python
lr_model.predict([[ 0.76696377, -0.75588528,  1.00904087,  0.30482333,  0.19028142]])
```

```python
lr_model.score(x_train_sd,y_train)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
with open('model.pkl','wb') as files:
  pickle.dump((lr_model,scaler,le),files)
```
