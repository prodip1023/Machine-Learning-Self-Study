# EXPLORATORY DATA ANALYSIS(EDA):
- All Machine learning models or algorithims work on data.When there is incorrect data or the data contains abnormal values or the data contains duplicate rows,then that data will lead to inaccurate conclusions from the Machine Learning model.Hence,it is needed that we have to analyse the data and clean it before we give it to a Machine Learning model.This is called EDA.

# THE PROCESS OF EDA:
- Knowing initial details about data.
- Modifying or removing unwanted data
- Retrieving data
- Getting Statistical information
- Drawing graphs/plots.

# Data Checks to perform

- Check Missing values
- Check Duplicates
- Check data type
- Check the number of unique values of each column
- Check statistics of data set
- Check various categories present in the different categorical column

#### Check the unique values in "Origin" Column

```python
# Find out how many continents are manufacturing the cars
cars['Origin'].unique()
```

#### Convert Continent names into numeric

```python
def Origin_to_Numeric(x):
    if x == 'Asia':
        return 1
    if x == 'Europe':
            return 2
    if x == 'USA':
        return 3
```

```python
# Now apply the above method on origin column
cars['Origin'] = cars['Origin'].apply(Origin_to_Numeric)
```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in cars.columns if cars[feature].dtype != 'O']
categorical_features = [feature for feature in cars.columns if cars[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# proportion of count data on categorical columns
for col in categorical_features:
    print(cars[col].value_counts(normalize=True) * 100)
    print('---------------------------')
```
